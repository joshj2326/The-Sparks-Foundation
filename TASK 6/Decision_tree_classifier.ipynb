# TASK 6 - Prediction using Decision Tree Algorithm

Create the Decision Tree classifier and visualize it graphically. 
The purpose is if we feed any new data to this classifier, it would be able to 
predict the right class accordingly. 


```python
# importing the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
%matplotlib inline 
```

```python
iris=load_iris()
iris.target
```

```python
iris_df=sns.load_dataset('iris')
iris_df.head()
```

```python
# independent and dependent features

X=iris_df.iloc[:,:-1]
y=iris.target
```

```python
# splitting into training and test set 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

# POSTPRUNING
dtreemodel= DecisionTreeClassifier()
```

```python
# training the model

dtreemodel.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(dtreemodel,filled=True)
```

From the above decision tree obtained after postpruning, at level 2 i.e value=[0,34,4] we see that the majority belongs to versicolor and for value=[0,1,30] the majority is viriginca. So,the unnecessary splitting is not required after level 2 as we know that the output would depend on the majority. Let's assign the max_depth parameter as 2. Note: These assumptions may not be true. The best way is by prepruning

```python
dtreemodel= DecisionTreeClassifier(max_depth=2)
dtreemodel.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(dtreemodel,filled=True)
```

```python
# prediciton

y_pred = dtreemodel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

The main aim of prepruning is to throw in all possible parameters and to check which combination gives the highest accuracy

```python
# PREPRUNING 

parameter= {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

dtreemodel=DecisionTreeClassifier()
cv=GridSearchCV(dtreemodel,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

Note: The above is not giving an error. It is experimenting the different combination of parameters to find the best one

```python
cv.best_params_
```

```python
y_pred=cv.predict(X_test)
y_pred
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```
