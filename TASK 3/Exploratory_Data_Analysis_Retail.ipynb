```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# importing the dataset

ss_data = pd.read_csv('/content/SampleSuperstore.csv')
ss_data.head()
```

```python
# describing the data

ss_data.describe()
```

```python
# checking for null values

print(ss_data.isnull().sum())
```

```python
# checking unique Segment values

ss_data.Segment.unique()
```

```python
# checking unique Region values

ss_data.Region.unique()
```

```python
# checking unique State values

ss_data.State.unique()
```

```python
# count value of states

print(ss_data['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=ss_data['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
# count value of region


print(ss_data['Region'].value_counts())
plt.figure(figsize=(6,6))
ss_data["Region"].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
# count value of category

print(ss_data['Category'].value_counts())

plt.figure(figsize=(6,6))
ss_data["Category"].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
# count value of quantity

print(ss_data['Quantity'].value_counts())
plt.figure(figsize=(6,6))
ss_data["Quantity"].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
# understanding the correlation

ss_data.corr()
```

```python
# rounding off the correlation values to 2 decimals

ss_data.corr().round(2)
```

```python
# correlation heatmaps

sns.heatmap(ss_data.corr().round(2),annot=True,vmin=-1,vmax=1,center=0,cmap='RdYlGn')
```

```python
matrix =ss_data.corr().round(2)
mask = np.triu(np.ones_like(matrix,dtype=bool))
mask
```

```python
# correlation heatmaps without the redundant values

sns.heatmap(matrix,annot=True,vmin=-1,vmax=1,center=0,cmap='RdYlGn',mask=mask)
```

Here, Sales has a strong positive correlation with the Profit earned as compared to the others. Also, discount has the highest negative correlation 

```python
ss_data.plot.scatter(x='Discount',y='Profit',s=5)
```

As the discount increases, the profit tends to reduce as show by the negative correlation in the matrix 

```python
ss_data.plot.scatter(x='Sales',y='Profit',s=5)
```

```python
plt.figure(figsize = (11,5))
sns.lineplot('Sales', 'Profit', data = ss_data, color = 'r', label= 'Profit')
plt.legend()
```

Increased sales leads to increased profit. The behavior is as expected from the positive correlation

```python
plt.figure(figsize = (11,5))
sns.lineplot('Segment', 'Sales', data = ss_data, color = 'r', label= 'Sales')
plt.legend()
```

```python
sns.barplot(x=ss_data.Segment,y=ss_data.Profit)
```

The Home Office Segment contributes the most to profit as it has more sales while the Consumer segment has a less profit due to lesser number of sales

```python
sns.barplot(x=ss_data.Category,y=ss_data.Discount)
```

Furniture provides the highest discount while technology provides the lowest 

```python
sns.barplot(x=ss_data.Category,y=ss_data.Profit)
```

Technology contibutes the most to profit as it provides the least discount. Furniture and office supplies does not contribute much to the profit

```python
sns.barplot(x=ss_data.Discount,y=ss_data.Profit)
```

As the discount increases the profit begins to reduce

```python
sns.barplot(x=ss_data.Region,y=ss_data.Sales)
```

All the regions contribute almost equally to the profit
